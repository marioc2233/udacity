```python
def list_files():
    list_files = listdir('../data/')
    final = []
    for file in list_files:
        print(len(file.split('.')))
        if len(file.split('.')) == 2:
            final.append(file)
            
    return final

def return_df_file(file):
    df = pd.read_csv(f'data/{file}', sep=';')
    return df
```
